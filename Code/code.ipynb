```python
import pandas as pd
import matplotlib.pyplot as plt


files = ['ราคา.xlsx', 'รีวิว.xlsx', 'โปรโมชั่น.xlsx', 'ทดลองอ่าน.xlsx', 
         'ตอนพิเศษเฉพาะ E-Book.xlsx', 'ชนิดไฟล์ EPUB.xlsx', 'อื่นๆ.xlsx']

factors = ['price', 'review', 'promotion', 'trial for free', 
           'Spacial Episode', 'Typefile_EPUB', 'other']


rows_count = []


for file in files:
    df = pd.read_excel(file)  # ใช้ read_excel สำหรับการอ่านไฟล์ .xlsx
    rows_count.append(len(df))  # นับจำนวนแถวด้วย len(df)


plt.bar(factors, rows_count, color='skyblue')


plt.xlabel('Factors of E-Book', fontsize=14)
plt.ylabel('Number', fontsize=14)
plt.title('E-Book', fontsize=16)

plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()


```

```python
import pandas as pd

input_file = 'ทดเลข.xlsx'
df = pd.read_excel(input_file)

# สร้าง mapping สำหรับกลุ่มคำและค่าที่ต้องการ return
keyword_map = {
    "สร้างแรงบันดาลใจ / สร้างความหวัง / สร้างพลังบวก": "How to",
    "เข้าใจตนเอง / เข้าใจผู้อื่น": "How to",
    "พัฒนาตนเอง / พัฒนาทักษะ / สร้างนิสัย / เปลี่ยนพฤติกรรม": "How to",
    "เข้าใจง่าย / กระชับ ตัดตรง / นำไปใช้ได้เลย": "Shortcut",
    "เคล็ดลับจากคนที่สำเร็จ / แนวทางสู่ความสำเร็จ": "Shortcut",
    "รู้ลึก / ลึกซึ้ง": "Sophia",
    "ปัญญาความรู้วิชาการ": "Sophia"
}

# ฟังก์ชันสำหรับการหาคำที่ตรงกับข้อความและ return ค่าที่กำหนด
def find_keywords(text):
    results = []
    keywords = text.split(';')  # แยกคำที่คั่นด้วย ;
    for keyword in keywords:
        keyword = keyword.strip()  # ลบช่องว่างรอบคำ
        for key, value in keyword_map.items():
            if key in keyword:
                results.append(value)
    return results

# สร้าง DataFrame ใหม่เพื่อเก็บผลลัพธ์
output_data = []

for index, row in df.iterrows():
    text = row['แนว']  # เปลี่ยนเป็นชื่อ column ที่ต้องการอ่าน
    results = find_keywords(text)
    if results:
        for result in results:
            output_data.append([text, result])

# สร้าง DataFrame จากผลลัพธ์
output_df = pd.DataFrame(output_data, columns=['Original Text', 'Return Value'])

# เขียนผลลัพธ์ลงไฟล์ Excel ใหม่
output_file = 'สำนักพิมพ์.xlsx'
output_df.to_excel(output_file, index=False)

print(f'สร้างไฟล์ {output_file} เสร็จเรียบร้อยแล้ว')
```

```python
import pandas as pd

# อ่านไฟล์ Excel รวมปัจจัย.xlsx
input_file = 'รวมปัจจัย.xlsx'
df = pd.read_excel(input_file)

# กำหนดคอลัมน์ที่ต้องการฟิลเตอร์ (เปลี่ยนชื่อคอลัมน์ให้ตรงกับไฟล์จริง)
column_name = 'ประเภทหนังสือที่ท่านชอบอ่านมากที่สุด'

# กำหนดประเภทหนังสือที่ต้องการฟิลเตอร์
accepted_types = ["Fiction (นิยาย)", "Essay (ความเรียง คำคม)", "Non Fiction (ความรู้, พัฒนาตนเอง)"]

# ฟังก์ชันในการแปลงประเภทหนังสือ
def transform_book_type(book_type):
    if book_type not in accepted_types:
        return "การ์ตูนและอื่นๆ"
    return book_type

# แปลงข้อมูลในคอลัมน์ประเภทหนังสือ
df[column_name] = df[column_name].apply(transform_book_type)

# เขียนผลลัพธ์ลงไฟล์ Excel ใหม่
output_file = 'ประเภทของ E-Book ที่สนใจ.xlsx'
df.to_excel(output_file, index=False)

print(f'สร้างไฟล์ {output_file} เสร็จเรียบร้อยแล้ว')
```

```python
import pandas as pd

# อ่านไฟล์ Excel
file_path = 'ช่วงเวลาที่เล่นโซเชี่ยว.xlsx'
df = pd.read_excel(file_path)

# เลือกคอลัมน์ A, B, C, D, F
columns = ['A', 'B', 'C', 'D', 'F']

# ละเว้นช่องว่าง และรวมข้อมูลทุกคอลัมน์ลงในคอลัมน์ A
combined_data = pd.concat([df[col].dropna() for col in columns], ignore_index=True)

# แสดงข้อมูลที่รวมแล้ว
print(combined_data)

# บันทึกเป็นไฟล์ใหม่ (ถ้าต้องการ)
combined_data.to_excel('combined_data.xlsx', index=False)

```
